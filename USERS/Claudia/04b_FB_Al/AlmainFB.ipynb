Aluminum(III) speciation with a single Al(OH)3(s) solid phase,
using the Fischer-Burmeister formulation (ONE solid phase).
It ignore e- and Cl- columns, so A_solid and K_solid are scalars.

Inputs:
   - AlT : total aluminum concentration (M)
   - pH range

Output:
   - results  : structure containing 1D data (pH, species, SI, etc.)
   - results2D: structure containing 2D “surface” data (pH vs. AlT)

This function also generates various plots (using results and -results2D)

Based on the original code by Claudia Santiviago Petzoldt
