```octave
% =========================
% Setup (graphics + helper)
% =========================
warning off;
close all;

toolkits = available_graphics_toolkits();

if any(strcmp(toolkits,"qt"))
  graphics_toolkit("qt");          % mejor si existe
elseif any(strcmp(toolkits,"fltk"))
  graphics_toolkit("fltk");
else
  graphics_toolkit("gnuplot");     % fallback
end

% Export bonito para gnuplot (cairo)
setenv("GNUTERM", "pngcairo size 1400,900 font ',12'");

set(0,"defaultlinelinewidth",2);
set(0,"defaultaxesfontsize",13);
set(0,"defaulttextfontsize",13);

% --- Helper robusto: guarda UNA figura (la actual) ---
function save_png(fname)
  if nargin < 1, fname = "figure.png"; end
  set(gcf,"paperpositionmode","auto");
  set(gcf,"color","w");
  drawnow();
  pause(0.05);  % deja respirar a gnuplot/jupyter

  if strcmp(graphics_toolkit(),"gnuplot")
    % OJO: es -dpngcairo (con N), no "dpmg..."
    print(gcf, fname, "-dpngcairo", "-r150");
  else
    print(gcf, fname, "-dpng", "-r150");
  end

  drawnow();
  pause(0.05);
end

% --- Helper: guarda TODAS las figuras abiertas (Fig1.png, Fig2.png, ...) ---
function save_all_figs(prefix)
  if nargin < 1, prefix = "Fig"; end
  figs = sort(allchild(0));    % handles de figuras
  figs = flipud(figs);         % Fig1, Fig2, ...
  for k = 1:numel(figs)
    figure(figs(k));
    save_png(sprintf("%s%d.png", prefix, k));
    close(gcf);                % importante: evita cuelgues y lentitud
  end
end
```

```octave
% -- Simulation Inputs --
pHvals = 2:0.2:12;  % pH range for the simulation
FeT    = 1e-5;      % Total Iron concentration (mol/L)
PT     = 1e-5;      % Total Phosphate concentration (mol/L)
peINT  = 5.75;      % pe0 value (redox condition)
HFOsi  = 0.005;     % Moles of strong adsorption sites per mole of Fe(OH)3(s)
HFOwi  = 0.2;       % Moles of weak adsorption sites per mole of Fe(OH)3(s)
```

```octave
adaptive_Fe_redox_adsorption_FB_solver6(pHvals, FeT, PT, peINT, HFOsi, HFOwi);
save_all_figs("Fig");
system("ls -lh Fig*.png");
```
