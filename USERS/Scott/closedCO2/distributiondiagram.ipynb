# distribution diagram for closed CO2 system

```octave
% constants

Ka1=10^-6.3; Ka2=10^-10.3; CT=10e-3; pH=3:0.1:12; H=10.^-pH;
```

```octave
denom=H.^2+H*Ka1+Ka1*Ka2*ones(size(pH));

alphaHCO3=(Ka1*H)./denom;
alphaCO3=Ka1*Ka2*ones(size(pH))./denom;
alphaH2CO3=(H.^2)./denom;
```

```octave
% graphing

graphics_toolkit ("plotly");

plot(pH,alphaH2CO3,pH,alphaHCO3,pH,alphaCO3)
legend('H2CO3','HCO3','CO3','location','eastoutside')

xlabel('pH'); ylabel('\alpha')
```

```octave
% graphing

graphics_toolkit ("plotly");

plot(pH,CT*alphaH2CO3,pH,CT*alphaHCO3,pH,CT*alphaCO3)
legend('H2CO3','HCO3','CO3','location','eastoutside')

xlabel('pH'); ylabel('[species]')
```
